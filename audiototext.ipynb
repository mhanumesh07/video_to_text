```python
import moviepy.editor as mp
from transformers import pipeline

# Step 1: Extract audio from the video
video_path = "/content/One of the Greatest Speeches Ever  Steve Jobs.mp4"  # Replace with the path to your video file
audio_path = "extracted_audio.wav"

video = mp.VideoFileClip(video_path)
audio = video.audio
audio.write_audiofile(audio_path)

# Step 2: Convert audio to text using the Hugging Face Transformers library
asr = pipeline("automatic-speech-recognition")

# Provide the path to the extracted audio file
audio_file = audio_path

# Convert audio to text
text = asr(audio_file)

# Extract the transcribed text
transcribed_text = text[0]['alternative'][0]['text']

# Step 3: Print the transcribed text
print("Transcribed text:", transcribed_text)
```
